## Criando um arquivo JSON

```python
import json

with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False)
```

### Lendo o JSON e criando o DataFrame

```python
import json

with open('data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame.from_dict(data)
df
```

## Medidas de Posição

```python
# media
df['idade'].mean()
```

```python
# mediana
df['idade'].median()
```

```python
# moda
df['idade'].mode()
```

## Medidas de Dispersão

```python
# variância
df['idade'].var()
```

```python
# desvio padrão
df['idade'].std()
```

```python
# coeficiente de variação
df['idade'].std() / df['idade'].mean() * 100
```

## Medidas de Forma

```python
# assimetria
df['idade'].skew()
```

```python
# curtose
df["idade"].kurtosis()
```

```python
# medidas estatísticas descritivas
df['idade'].describe()
```

## Correlação de variáveis

```python
# Correlação segundo Pearson
df.select_dtypes(include='number').corr(method='pearson')
```

```python
# Correlação segundo Spearman
df.select_dtypes(include='number').corr(method='spearman')
```

```python
# Correlação entre duas variaveis
df.idade.corr(df.altura)
```
